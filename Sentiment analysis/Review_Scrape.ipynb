```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
import time
import random
import json
import csv
import requests
import json
import urllib.request, urllib.error, urllib.parse
import pprint
import nltk
nltk.download('all')
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
urls_gap = []
urls_zara = []
for x in range(1,16) :
    urls_gap.append('https://www.sitejabber.com/reviews/gap.com?page=' + str(x) + '&sort=published#reviews')
    urls_zara.append('https://www.sitejabber.com/reviews/zara.com?page=' + str(x) + '&sort=published#reviews')
```

```python
firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)
text_gap = []
text_zara = []

for url in urls_gap:
    driver.get(url) 
    time.sleep(random.uniform(3,7))

    html = driver.page_source 
    soup = bs(html, "html.parser")
    for html_class in soup.find_all('div', class_="review__content"):
        sublink = html_class.find_all('review__text')
        sublink_text = html_class.find('div', class_='review__text').get_text()
        sublink_text = sublink_text.replace('\n','')
        text_gap.append(sublink_text)

        
for url in urls_zara:
    driver.get(url) 
    time.sleep(random.uniform(3,7))

    html = driver.page_source 
    soup = bs(html, "html.parser")
    for html_class in soup.find_all('div', class_="review__content"):
        sublink = html_class.find_all('review__text')
        sublink_text = html_class.find('div', class_='review__text').get_text()
        sublink_text = sublink_text.replace('\n','')
        text_zara.append(sublink_text)

print(text_zara)
driver.quit() 
sentence_list = []
word_list = []
for i in sublink_text:
    sentence_list = sent_tokenize(i)
for x in sentence_list:
    word_list = word_tokenize(x)
    
```
