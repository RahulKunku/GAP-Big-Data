## Pytrends in Python

Let's use pytrends in python to scrape Google Trends data.

```python
pip install pytrends
```

```python
import pandas as pd
from pytrends.request import TrendReq
import time

# Initialize pytrends
pytrends = TrendReq()

# Define your seasons with date range and names
season_dates = {
    'Spring': '2022-03-01 2022-05-31',
    'Summer': '2022-06-01 2022-08-31',
    'Fall': '2022-09-01 2022-11-30',
    'Winter': '2022-12-01 2023-02-28'
}

categories = ['clothing', 'blouses', 'denim', 'dresses', 'pants', 'shirts', 'shorts', 'skirts', 'sweaters', 'sweatshirts']

dataframes = {}

for category in categories:
    kw_list = [category]
    for season, date_range in season_dates.items():
        print(f"Fetching data for {kw_list}, {season}")
        try:
            pytrends.build_payload(kw_list, cat=0, timeframe=date_range, geo='US')

            data_topics = pytrends.related_topics()
            data_queries = pytrends.related_queries()

            # Store data in dictionary with new naming convention
            dataframes[f"{category}_{season}_topics_top"] = pd.DataFrame(data_topics[category]['top'])
            dataframes[f"{category}_{season}_topics_rising"] = pd.DataFrame(data_topics[category]['rising'])
            dataframes[f"{category}_{season}_queries_top"] = pd.DataFrame(data_queries[category]['top'])
            dataframes[f"{category}_{season}_queries_rising"] = pd.DataFrame(data_queries[category]['rising'])

        except Exception as e:
            print(f"Failed to fetch data for {kw_list}, {season}. Error: {e}")

            if "429" in str(e):
                print("We're being rate-limited. Sleeping for longer...")
                time.sleep(60)  # Sleep for 1 minute

            continue

        time.sleep(30)  # Add a delay to prevent rate limiting

# Example: Access and print a specific DataFrame
print(dataframes.get('clothing_Spring_queries_top', "DataFrame not found"))
```

## Related Topics

Users searching for your term also searched for these topics. You can view by the following metrics:
* Top - The most popular topics. Scoring is on a relative scale where a value of 100 is the most commonly searched topic and a value of 50 is a topic searched half as often as the most popular term, and so on.

* Rising - Related topics with the biggest increase in search frequency since the last time period. Results marked "Breakout" had a tremendous increase, probably because these topics are new and had few (if any) prior searches.

```python
df_Spring_top = pd.DataFrame()
df_Spring_rising = pd.DataFrame()

df_Summer_top = pd.DataFrame()
df_Summer_rising = pd.DataFrame()

df_Fall_top = pd.DataFrame()
df_Fall_rising = pd.DataFrame()

df_Winter_top = pd.DataFrame()
df_Winter_rising = pd.DataFrame()


for category in categories:
    for season, date_range in season_dates.items():
        df_top = dataframes[f"{category}_{season}_topics_top"][['value','topic_title']]
        df_top['category'] = category
        df_top['season'] = season
        
        if season == 'Spring':
            df_Spring_top = pd.concat([df_Spring_top, df_top])
        elif season == 'Summer':
            df_Summer_top = pd.concat([df_Summer_top, df_top])
        elif season == 'Fall':
            df_Fall_top = pd.concat([df_Fall_top, df_top])
        else:
            df_Winter_top = pd.concat([df_Winter_top, df_top])

            
for category in categories:
    for season, date_range in season_dates.items():
        key = f"{category}_{season}_topics_rising"
        
        # Check if the key exists in the dataframes dictionary
        if key not in dataframes:
            print(f"Data for {key} does not exist!")
            continue

        # Check if 'value' and 'topic_title' columns exist
        if 'value' not in dataframes[key].columns or 'topic_title' not in dataframes[key].columns:
            print(f"Columns are missing in {key}. Actual columns are: {dataframes[key].columns}")
            continue

        df_rising = dataframes[key][['value', 'topic_title']]
        df_rising['category'] = category
        df_rising['season'] = season

        if season == 'Spring':
            df_Spring_rising = pd.concat([df_Spring_rising, df_rising])
        elif season == 'Summer':
            df_Summer_rising = pd.concat([df_Summer_rising, df_rising])
        elif season == 'Fall':
            df_Fall_rising = pd.concat([df_Fall_rising, df_rising])
        else:
            df_Winter_rising = pd.concat([df_Winter_rising, df_rising])
            

# Save the aggregated dataframes to CSV files
df_Spring_top.to_csv('topic_Spring_top.csv', index=False)
df_Spring_rising.to_csv('topic_Spring_rising.csv', index=False)

df_Summer_top.to_csv('topic_Summer_top.csv', index=False)
df_Summer_rising.to_csv('topic_Summer_rising.csv', index=False)

df_Fall_top.to_csv('topic_Fall_top.csv', index=False)
df_Fall_rising.to_csv('topic_Fall_rising.csv', index=False)

df_Winter_top.to_csv('topic_Winter_top.csv', index=False)
df_Winter_rising.to_csv('topic_Winter_rising.csv', index=False)
        
```

```python
# Concatenate the "top" dataframes for all seasons
df_all_seasons_top = pd.concat([
    df_Spring_top, 
    df_Summer_top, 
    df_Fall_top, 
    df_Winter_top
], ignore_index=True)

# Save to CSV
df_all_seasons_top.to_csv('all_seasons_top_topics.csv', index=False)

# Concatenate the "rising" dataframes for all seasons
df_all_seasons_rising = pd.concat([
    df_Spring_rising, 
    df_Summer_rising, 
    df_Fall_rising, 
    df_Winter_rising
], ignore_index=True)

# Save to CSV
df_all_seasons_rising.to_csv('all_seasons_rising_topics.csv', index=False)
```
