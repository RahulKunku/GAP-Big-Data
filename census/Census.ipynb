```python
import requests
import pandas as pd

base_url = "https://www2.census.gov/programs-surveys/popest/tables/2020-2022/state/detail/sc-est2022-agesex-{:02d}.xlsx"

for i in range(1, 57):  # Loop from 1 to 56
    url = base_url.format(i)  # Format the URL with the current index

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception if the HTTP response is not successful

        # Extract the state name from the text
        data = pd.read_excel(url)
        title = str(data.iloc[0, 0])  # Convert the value to a string
        
        # Find the starting and ending positions for extraction
        start_pos = title.find("Age Groups by Sex for ") + len("Age Groups by Sex for ")
        end_pos = title.find(":", start_pos)

        # Extract the state name
        if start_pos == -1 or end_pos == -1:
            print(f"Error: Could not find the expected substrings in the title for file {i}.")
            state_name = f"default_{i}"
        else:
            state_name = title[start_pos:end_pos].strip()

        filename = f"{state_name}.xlsx"

        # Save the file
        with open(filename, "wb") as f:
            f.write(response.content)

        print(f"File saved as {filename}")

    except requests.HTTPError:  # Handle HTTP errors
        print(f"File not found for URL: {url}")
        continue

print("All files processed!")
```
